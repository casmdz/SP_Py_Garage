````python
class ParkingGarage:
    def __init__(self):
        self.tickets = 10
        self.parking_spaces = 10
        self.current_ticket = {}

    def takeTicket(self):
        if self.tickets > 0 and self.parking_spaces > 0:
            self.tickets -= 1
            self.parking_spaces -= 1
            ticket_number = len(self.current_ticket) + 1
            self.current_ticket[ticket_number] = {
                "paid": False}
            print(f"This is your ticket, #{ticket_number}")
            return ticket_number
        #uh
        else:
            print("No parking available")

    def payForParking(self):
#         ticket_number = int(input("Enter your ticket number to pay: "))
        while True:
            ticket_number = int(input("Enter your ticket number to pay: "))
            if ticket_number in self.current_ticket and not self.current_ticket[ticket_number]['paid']:
                payment = input("You are at the paybox. Enter '$5': ")
                if payment:
                    self.current_ticket[ticket_number]['paid'] = True
                    print('Your ticket has been paid at the paybox. You have 15 minutes to leave.')
#                     break
                else:
                    print("You have proceeded without paying at payForParking")
# #                     print('Payment not received, please try again')
                break
            else:
                print("Invalid ticket number. Please try again.")
#             print("try again ?? Invalid ticket number at paybox")
#             #print("You have proceeded without paying at payForParking")

    def leaveGarage(self):
        while True:
            ticket_number = int(input("You are at the exit. Enter your ticket number: "))
            if ticket_number in self.current_ticket:
    #             while True:
                if self.current_ticket[ticket_number]['paid']:
                    self.tickets += 1
                    self.parking_spaces += 1
                    del self.current_ticket[ticket_number]
                    print('Thank you for paying the correct ticket. Have a nice day')
                    break
                else:
                    while True:
                        payment = input('Payment not received at box, please try here. Enter payment of "$5": ')
                        if payment:
                            self.current_ticket[ticket_number]['paid'] = True
                            print('Your paid your ticket, you may leave now!!')
                            break
                        else:
                            print('Payment not received at garage, please try again')
#                 break
            else: 
                print("Invalid ticket number at leaveGarage")                   


garage = ParkingGarage()

garage.takeTicket()
garage.payForParking()
garage.leaveGarage()
````
